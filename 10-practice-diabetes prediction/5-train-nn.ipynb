```python
import numpy as np
import torch
import torch.nn as nn
import pandas as pd
from sklearn.preprocessing import StandardScaler
from torch.utils.data import Dataset, DataLoader
```

```python
# Load the data
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
# For x extract all columns except the last one
x = df.iloc[:, 0:-1].values
# For y extract the last column
y_string = list(df.iloc[:, -1])
```

```python
x
```

```python
y_string
```

```python
# Convert the y_string to y integer
y_int = []
for i in y_string:
    if i == 'positive':
        y_int.append(0)
    else:
        y_int.append(1)

y_int
```

```python
# Convert the y_int to an array
y = np.array(y_int, dtype=np.float64)
y
```

```python
# Feature Normalization
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

```python
# Convert the x and y to tensors
x = torch.tensor(x)
y = torch.tensor(y).unsqueeze(1)
x.shape, y.shape
```

```python
class DiabetesDataset(Dataset):
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.n_samples = x.shape[0]
        
    def __getitem__(self, index):
        return self.x[index], self.y[index]
    
    def __len__(self):
        return self.n_samples
    
```

```python
# Create the dataset using the class DiabetesDataset
dataset = DiabetesDataset(x, y)
len(dataset)
```

```python
# Load the dataset using the dataloader
train_loader = DataLoader(dataset=dataset, batch_size=32, shuffle=True)
```

```python
# visualize the train_loader data
print(f"Number of batches: {len(train_loader)}")
for x, y in train_loader:
    print(f"Data: {x.shape}, Labels: {y.shape}")
    break
```

```python
# Create the model
class DiabetesModel(nn.Module):
    def __init__(self, input_features):
        super(DiabetesModel, self).__init__()
        self.fc1 = nn.Linear(input_features, 5)
        self.fc2 = nn.Linear(5, 4)
        self.fc3 = nn.Linear(4, 3)
        self.fc4 = nn.Linear(3, 1)
        self.sigmoid = nn.Sigmoid()
        self.tanh = nn.Tanh()

    def forward(self, x):
        out = self.fc1(x)
        out = self.tanh(out)
        out = self.fc2(out)
        out = self.tanh(out)
        out = self.fc3(out)
        out = self.tanh(out)
        out = self.fc4(out)
        out = self.sigmoid(out)
        return out
```

```python
# Create the model object
model = DiabetesModel(x.shape[1])
model

# Create the loss function
criterion = nn.BCELoss(reduction='mean')

# Create the optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.9)


```

```python
# Train the model
num_epochs = 200
for epoch in range(num_epochs):
    for x, y in train_loader:
        x = x.float()
        y = y.float()

        # Forward pass
        y_pred = model(x)
        loss = criterion(y_pred, y)
        
        # Backward pass
        loss.backward()
        
        # Update the parameters
        optimizer.step()
        
        # Zero the gradients
        optimizer.zero_grad()
        
    if (epoch+1) % 10 == 0:
        print(f"Epoch: {epoch+1}, Loss: {loss.item():.4f}")

    # Accuracy calculation
    output = (y_pred > 0.5).float()
    accuracy = (output == y).float().mean()
    print(f"Accuracy: {accuracy.item():.4f}")
```
